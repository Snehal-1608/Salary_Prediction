```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df["workclass"]=le.fit_transform(df["workclass"])
```

```python
df["workclass"]
```

```python
df["marital-status"]=le.fit_transform(df["marital-status"])
```

```python
df["occupation"]=le.fit_transform(df["occupation"])
df["relationship"]=le.fit_transform(df["relationship"])
df["race"]=le.fit_transform(df["race"])
df["gender"]=le.fit_transform(df["gender"])
df["native-country"]=le.fit_transform(df["native-country"])
```

```python
x=df.drop(columns=["income"]) #x is my input means independent variable
y=df["income"]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=True, stratify=y)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier()
classifier.fit(xtrain,ytrain)
```

```python
y_pred=classifier.predict(xtest)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest,y_pred)
```

```python
def prediction(age, workclass, fnlwgt, educational_num, marital_status, occupation, relationship, race, gender, capital_gain, capital_loss, hours_per_week, native_country):
    prediction=classifier.predict([[age,workclass,fnlwgt,educational_num,marital_status,occupation,relationship,race,gender,capital_gain,capital_loss,hours_per_week,native_country]])
    print(prediction)
    return prediction
```

```python
def main():
    st.title("Salary Prediction")
    html_temp="""
    <div style ="background-color:green;padding:13px"> 
    <h1 style ="color:black;text-align:center;">Streamlit Salary Prediction Classifier ML App </h1> 
    </div> 
    """
    st.markdown(html_temp, unsafe_allow_html = True)
    age = st.text_input("Age", "Type Here") 
    workclass = st.text_input("Work Class", "Type Here") 
    fnlwgt = st.text_input("fnlwgt", "Type Here") 
    educational_num = st.text_input("Educational_num", "Type Here") 
    marital_status = st.text_input("Marital Status", "Type Here") 
    occupation = st.text_input("Occupation", "Type Here") 
    race = st.text_input("Race", "Type Here") 
    gender = st.text_input("Gender", "Type Here") 
    capital_gain = st.text_input("Capital_gain", "Type Here") 
    capital_loss = st.text_input("Capital_loss", "Type Here") 
    hours_per_week = st.text_input("Hours Per Week", "Type Here")
    native_country = st.text_input("Native Country", "Type Here")
    result =""
    if st.button("Predict"):
        result=prediction(age, workclass, fnlwgt, educational_num, marital_status, occupation, relationship, race, gender, capital_gain, capital_loss, hours_per_week, native_country)
    st.success("The output is {}".format(result))  
    
if __name__=='__main__':
    main()
```
